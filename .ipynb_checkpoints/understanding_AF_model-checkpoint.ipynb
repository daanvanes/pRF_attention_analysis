```python
%pylab inline
from AFSimulations import *
from __future__ import division
import seaborn as sn
```

```python
AF_dir = os.path.join('/Users/daanvanes/disks/Aeneas_Shared/2015/visual/PRF_2/AF_simulations/')
```

```python
fix_RF_x = 0.2
fix_RF_y = 0
fix_RF_size = 1

fix_AF_size = 2
bar_AF_size = 1.1

res = 131
maif = 3
stim_radius = 3.6
```

say we have a given pRF at fixation

```python

fix_RF = twodgauss(xo=fix_RF_x, yo=fix_RF_y, sigma_x=fix_RF_size,max_eccentricity=stim_radius,res=res,circle_mask=True)
imshow(fix_RF,origin='lowerleft')
title('pRF for Attend Fixation')
axis('off');
```

include AF at fixation: 

```python
fix_AF = twodgauss(xo=0, yo=0, sigma_x=fix_AF_size,max_eccentricity=stim_radius,res=res,circle_mask=True)

subplot(131)
imshow(fix_RF,origin='lowerleft')
title('pRF Fixation')
axis('off');
subplot(132)
imshow(fix_AF,origin='lowerleft')
title('AF')
axis('off');
subplot(133)
SD = (fix_RF/fix_AF)
SD[np.isnan(SD)] = 0
imshow(SD,origin='lowerleft')
title('SD')
axis('off');
```

let's first create AFs at the bar

```python
AFs = convolve_bar_with_AF(AF_size_intercept=bar_AF_size,AF_size_slope=0,model_area_increase_factor=maif,res=res,plot_dir=AF_dir,stim_radius=stim_radius,circle_mask=True,AF_surround_amp=0,AF_surround_ratio=1)[0]
fix_AF = twodgauss(xo=0, yo=0, sigma_x=fix_AF_size,max_eccentricity=stim_radius*maif,res=res,circle_mask=True)
fix_RF = twodgauss(xo=fix_RF_x, yo=fix_RF_y, sigma_x=fix_RF_size,max_eccentricity=stim_radius*maif,res=res,circle_mask=True)
SD = fix_RF/fix_AF
SD[np.isnan(SD)] = 0

# # now simulate attention to bar effect on this SD:
bar_RF = np.mean([(SD*AF)/np.max(SD*AF) for AF in AFs],axis=0)
bar_RF[disk(int((res-1)/2))==0] = 0    

# find maximum position
x_view = sp.signal.resample(np.max(bar_RF,axis=0),res*10)
y_view = sp.signal.resample(np.max(bar_RF,axis=1),res*10)
predicted_pos = np.argmax(x_view)/10, np.argmax(y_view)/10

# find maximum position
x_view = sp.signal.resample(np.max(fix_RF,axis=0),res*10)
y_view = sp.signal.resample(np.max(fix_RF,axis=1),res*10)
pRF_fix_pos = np.argmax(x_view)/10, np.argmax(y_view)/10
```

```python
figure()
subplot(131)
imshow(fix_RF[44:-44,44:-44],origin='lowerleft')
title('pRF Fixation')
axis('off');
subplot(132)
imshow(SD[44:-44,44:-44],origin='lowerleft')
title('SD')
axis('off');
subplot(133)
imshow(bar_RF[44:-44,44:-44],origin='lowerleft')
title('predicted PRF Stimulus')
axis('off');

figure()
subplot(aspect='equal')
pRF_fix_pos = np.array(pRF_fix_pos)-44
predicted_pos = np.array(predicted_pos)-44
plot(pRF_fix_pos[0],pRF_fix_pos[1],'o',label='pR Fix')
plot(predicted_pos[0],predicted_pos[1],'o',label='pRF Stim')
legend();

arrow(pRF_fix_pos[0],pRF_fix_pos[1],predicted_pos[0]-pRF_fix_pos[0],predicted_pos[1]-pRF_fix_pos[1])
axhline(43/2)
axvline(43/2)
xlim(0,43)
ylim(0,43)
xticks([]);
yticks([]);

print 'ecc diff: %.2f'%(linalg.norm(predicted_pos)-linalg.norm(pRF_fix_pos))
```

simulating the effect for area V3

```python
random_eccs = np.abs(random.normal(0,stim_radius*0.3,200))
valid_pRFs = (np.abs(random_eccs)<stim_radius)
random_eccs = random_eccs[valid_pRFs]
print n_pRFs.sum()
n_pRFs = valid_pRFs.sum()
random_polars = random.uniform(-np.pi,np.pi,n_pRFs)
random_xs = random_eccs * np.cos(random_polars)
random_ys = random_eccs * np.sin(random_polars)
random_sizes = 0.5+0.25*random_eccs

figure(figsize=(18,9))
subplot(131,aspect='equal')
plot(random_xs,random_ys,'o')
xlim(-stim_radius,stim_radius)
ylim(-stim_radius,stim_radius)

subplot(132)
hist(random_eccs)
subplot(133)
hist(random_sizes)

```

```python
pRF_fix_poss = []
pRF_stim_poss = []
SD_poss = []
for i in range(n_pRFs):
    x = random_xs[i]
    y = random_ys[i]
    s = random_sizes[i]
    fix_RF = twodgauss(xo=x, yo=y, sigma_x=s,max_eccentricity=stim_radius*maif,res=res,circle_mask=True)
    SD = fix_RF/fix_AF
    SD[np.isnan(SD)] = 0

    # # now simulate attention to bar effect on this SD:
    bar_RF = np.mean([(SD*AF)/np.max(SD*AF) for AF in AFs],axis=0)
    bar_RF[disk(int((res-1)/2))==0] = 0    

    pRF_stim_poss.append(find_profile_maxval(bar_RF,stim_radius*maif,use_circle_mask=True,upsample_factor=10))
    pRF_fix_poss.append(find_profile_maxval(fix_RF,stim_radius*maif,use_circle_mask=True,upsample_factor=10))
    SD_poss.append(find_profile_maxval(SD,stim_radius*maif,use_circle_mask=True,upsample_factor=10))



```

```python
from GroupLevelPlotting import GroupLevelPlots
from GroupLevelPlotting import General_functions

GF = General_functions()


predicted_fix_eccs = linalg.norm(pRF_fix_poss,axis=1)
predicted_stim_eccs = linalg.norm(pRF_stim_poss,axis=1)
SD_eccs = linalg.norm(SD_poss,axis=1)
from scipy.stats import binned_statistic
    
y = binned_statistic(SD_eccs, predicted_fix_eccs-SD_eccs, bins=4)[0]
x = binned_statistic(SD_eccs, SD_eccs, bins=4)[0]

n_bins = 6

color1 = colorsys.hsv_to_rgb(0,0,0.75)
color2 = colorsys.hsv_to_rgb(0,0,0.5)
color3 = colorsys.hsv_to_rgb(0,0,0.25)


figure(figsize=(6,3))
subplot(121)
# plot(SD_eccs,predicted_fix_eccs-SD_eccs,'o',c=color1,label='Fixation pRF-SD',alpha=0.1)
y = binned_statistic(SD_eccs, predicted_fix_eccs-SD_eccs, bins=n_bins)[0]
x = binned_statistic(SD_eccs, SD_eccs, bins=n_bins)[0]
plot(x,y,c=color1,lw=2)
# plot(SD_eccs,predicted_stim_eccs-SD_eccs,'o',c=color1,label='Stimulus pRF - SD',alpha=0.1)
y = binned_statistic(SD_eccs, predicted_stim_eccs-SD_eccs, bins=n_bins)[0]
x = binned_statistic(SD_eccs, SD_eccs, bins=n_bins)[0]
ylims = GF.find_yticks(gca().get_ylim(),ndec=2)
yticks(ylims[0],ylims[1])
plot(x,y,c=color2,lw=2)
xticks([0,3.6])
axhline(0,c='k')

xlabel('SD ecc')
ylabel(r'$\Delta$ ecc')
subplot(122)

sn.despine(offset=2)
# plot(predicted_fix_eccs,predicted_stim_eccs-predicted_fix_eccs,'o',c=color3,alpha=0.1)
y = binned_statistic(predicted_fix_eccs, predicted_stim_eccs-predicted_fix_eccs, bins=n_bins)[0]
x = binned_statistic(predicted_fix_eccs, predicted_fix_eccs, bins=n_bins)[0]
axhline(0,c='k')
plot(x,y,c=color3,lw=2)
xlabel('fix ecc')
# ylabel('stim-fix ecc')
ylims = GF.find_yticks(gca().get_ylim(),ndec=2)
yticks(ylims[0],ylims[1])
tight_layout()
xticks([0,3.6])


savefig('/Users/daanvanes/disks/Aeneas_Shared/2015/visual/PRF_2/data/_group_level/plots/Figure_5/understanding_AF_V3.pdf')

```
